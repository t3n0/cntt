```python
# In this notebook we will perform a simple dft calculation on a very coarse k-grid
# and compare the results with the tight binding method for the same CNT

from cntt.swcnt import Swcnt

mycnt = Swcnt(4,2)

# define a coarse k-grid
mycnt.calculateCuttingLines(5)

# launch the dft calculation
# NOTE: remember to specify the pseudopotential directory
# can be both absolute path and relative to the working dir
pseudo_dir = 'path/to/your/pseudo_dir'
mycnt.calculateElectronBands('DFT', 'myDFTrough', nprocs = 4, pseudo_dir=pseudo_dir)

# launch the TB calculation
mycnt.calculateElectronBands('TB', 'myTB', sym = 'hel', gamma = 3.0, fermi = 0.0)

# position of k valleys for the exciton calculation
mycnt.calculateKpointValleys()

# excitons generated by the two different underlying electron structures
mycnt.calculateExcitonBands('EM','myTB', deltaK=10, bindEnergy=0.2)
mycnt.calculateExcitonBands('EM','myDFTrough', deltaK=10, bindEnergy=0.2)

mycnt.calculateDOS('electron', name = 'all', enSteps = 100)
mycnt.calculateDOS('exciton', enSteps = 100)

mycnt.plot()
```
